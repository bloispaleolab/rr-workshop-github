### Who had the biggest increase in population?
#### Think through the analytical approach

1. One approach: just use the years 1952 and 2007, excluding countries with only data from those two years
2. Another approach: use the min and max years for each country, then calculate the population difference
3. A third approach: fit a line to the temporal data and examine the slopes

Let's use approach 2:
- Create a new data frame that has region, pop from min year, and pop from max year.
- Then subtract pop in min year from pop in max year and divide by min year population
- Then sort the output in descending order, so the top of the list is our answer


# Plotting

```python
# Import packages for plotting
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
plt.hist(gapminder_copy['life Exp'])
plt.xlabel('life Exp')
plt.ylabel('count')
```

```python
sns.boxplot(x= 'year', y= 'life Exp', data=gapminder_copy)
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
```

```python
plt.scatter(gapminder_copy.gdpPercap, gapminder_copy['life Exp'])
plt.xlabel('gdpPercap')
plt.ylabel('life Exp')
```

```python
plt.scatter(gapminder_copy.gdpPercap, gapminder_copy['life Exp'])
plt.xscale('log')
plt.xlabel('gdpPercap')
plt.ylabel('life Exp')
plt.savefig('../log_plot.png')
```
